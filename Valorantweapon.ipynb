**Pie Chart**

```python
data_origin = dataset["Wall Penetration"].value_counts().to_dict()

data_pie = data_origin.values()
label_pie = data_origin.keys()
colors = sns.color_palette('pastel')[0:5]
```

```python
plt.pie(x = data_pie,
        labels = label_pie,
        colors = colors,
        autopct='%.0f%%'
)
plt.title("Wall Penetration")
plt.show()
```
